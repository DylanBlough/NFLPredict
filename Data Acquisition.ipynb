```python
import pandas as pd

pd.set_option('display.max_rows', 100)
pd.set_option('display.max_columns', 300)
```

```python
YEARS = [2019, 2020]
# uncomment for one season
# pd.read_csv('https://github.com/guga31bb/nflfastR-data/blob/master/data/' \
                         # 'play_by_play_' + str(YEAR) + '.csv.gz?raw=True',
                         # compression='gzip', low_memory=False)

data = pd.DataFrame()

for i in YEARS:  
    i_data = pd.read_csv('https://github.com/guga31bb/nflfastR-data/blob/master/data/' \
                         'play_by_play_' + str(i) + '.csv.gz?raw=True',
                         compression='gzip', low_memory=False)

    data = data.append(i_data, sort=True)

data.reset_index(drop=True, inplace=True) # reset index for each row

data.to_csv('./Datasets/stats.csv.gz', compression='gzip', index=False)
```

```python
# read data back in

data = pd.read_csv('./Datasets/stats.csv.gz', compression = 'gzip', low_memory = False)
```

```python
# cleaning
# removing columns that have been replaced
data.drop(['passer_player_name', 'passer_player_id',
           'rusher_player_name', 'rusher_player_id',
           'receiver_player_name', 'receiver_player_id'],
          axis=1, inplace=True)

# removes postseason
data = data.loc[data.season_type=='REG']

# remove kickoffs punts kneels downs spikes
data = data.loc[(data.play_type.isin(['no_play','pass','run'])) & (data.epa.isna()==False)]

# fix labeling of scrambles
data.play_type.loc[data['pass']==1] = 'pass'
data.play_type.loc[data.rush==1] = 'run'

# reset the index again
data.reset_index(drop=True, inplace=True)

# save the cleaned data
data.to_csv('./Datasets/stats_clean.csv.gz', compression = 'gzip', index = False)
```
